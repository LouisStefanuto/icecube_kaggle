```python
import numpy as np
import pandas as pd
```

# Split train meta into smaller files

```python
NUM_BATCHES = 659
```

```python
meta_train = pd.read_parquet('/kaggle/input/icecube-neutrinos-in-deep-ice/train_meta.parquet')
meta_train.head()
```

```python
batch_ids = meta_train.batch_id.unique()[:NUM_BATCHES]
batch_ids
```

```python
for batch_id in batch_ids:
    meta_train_i = meta_train[meta_train.batch_id == batch_id]
    meta_train_i.to_parquet(f'train_meta_{batch_id}.parquet')
```

# Create a smaller validation set - Meta and Batch

```python
BATCH_VAL_ID = 660
NUM_EVENTS = 20000
```

```python
meta_val = meta_train[meta_train.batch_id == BATCH_VAL_ID]
meta_val
```

```python
meta_val_small = meta_val.iloc[:NUM_EVENTS]
meta_val_small
```

```python
val_batch_df = pd.read_parquet(f'/kaggle/input/icecube-neutrinos-in-deep-ice/train/batch_{BATCH_VAL_ID}.parquet')
val_batch_df
```

```python
last_pulse_idx = meta_val_small.last_pulse_index.to_numpy()[-1]
val_batch_df = val_batch_df.iloc[:last_pulse_idx]
val_batch_df
```

```python
meta_val_small.to_parquet(f'val_meta_{BATCH_VAL_ID}_small.parquet')
val_batch_df.to_parquet(f'batch_{BATCH_VAL_ID}_small.parquet')
```
